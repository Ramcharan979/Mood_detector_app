```python
#  emotion classification pipeline
from transformers import pipeline

# Load model (simpler format, no top_k to avoid nested output)
emotion_classifier = pipeline(
    'text-classification',
    model='j-hartmann/emotion-english-distilroberta-base',
    revision='main')

def mood_to_emoji(mood_label):
    emoji_map = {
        "joy": "ğŸ˜„",
        "sadness": "ğŸ˜¢",
        "anger": "ğŸ˜ ",
        "fear": "ğŸ˜¨",
        "disgust": "ğŸ¤¢",
        "surprise": "ğŸ˜²",
        "love": "ğŸ¥°",
        "neutral": "ğŸ˜"
    }
    return emoji_map.get(mood_label.lower(), "â“")

def detect_text_emotion(text):
    try:
        if not isinstance(text, str) or len(text.strip()) < 2:
            print(f"âš ï¸ Skipping invalid or too-short input: '{text}'")
            return "unknown"

        result = emotion_classifier(text)
        label = result[0]['label']
        emoji = mood_to_emoji(label)
        print(f"âœ… Text: '{text}' â†’ Mood: {label} {emoji}")
        return f"{label} {emoji}"
    except Exception as e:
        print(f"âŒ Error: {e} for text: '{text}'")
        return "unknown"
```

```python
from deepface import DeepFace
def detect_image_emotion(image_path):
    analysis = DeepFace.analyze(img_path=image_path, actions=['emotion'], enforce_detection=False)
    emotion = analysis[0]['dominant_emotion']
    return emotion
```

## Demo: Text Emotion Detection with some given inputs texts

```python
sample_texts = [
    "I am so happy today!",
    "Why does everything go wrong...",
    "This is exciting!",
    "I'm scared of the future.",
    "Hi",
    ""  # Empty
]

for text in sample_texts:
    mood = detect_text_emotion(text)
    print(f"â¡ï¸ Result: {mood}")

```
